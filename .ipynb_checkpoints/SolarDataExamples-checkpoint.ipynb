```python
from astropy.io.fits import getdata, getheader
import matplotlib.pyplot as plt
%matplotlib inline


better_size = 16

plt.rc('font', size=better_size)          # controls default text sizes
plt.rc('axes', titlesize=better_size)     # fontsize of the axes title
plt.rc('axes', labelsize=better_size)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=better_size)    # fontsize of the tick labels
plt.rc('ytick', labelsize=better_size)    # fontsize of the tick labels
plt.rc('text', usetex=False)
```

```python
data, hdr = getdata('nu20211002001_sun_centerA_sr.pha', 1, header=True)

bgd, bgd_hdr = getdata('bgdnu20211002001_sun_centerA_sr.pha', 1, header=True)

# Scale factor here accounts for both the exposure and the amount of the detector that you're using.
scale = (hdr['EXPOSURE'] * hdr['BACKSCAL']) / (bgd_hdr['EXPOSURE'] * bgd_hdr['BACKSCAL'])

```

# Plot is Counts / Channel

```python
plt.figure(figsize=(15, 10))
plt.loglog(data['CHANNEL'], data['COUNTS'], '+')
plt.loglog(data['CHANNEL'], bgd['COUNTS']*scale)
plt.ylim([1e-4, 1000])
plt.ylabel("Counts")
plt.xlabel("Channel")
plt.axvline(x=23, linestyle='dashed')

```

## Okay, so here the blue data points are the data (in counts per channel) and the gold are the expected counts per channel due to the background. The dashed vertical line is channel 23 (or 2.52 keV).

## Note that I've jacked up the effective exposure in the background by a factor of 1e6 to get a "smooth" background model. I've also turned off the "solar" background components in the model.

## We generate this using nuskybgd, which we run *after* we've extracted the source spectrum using nuproducts. The resulting PHA files have all of the keywords necessary for appropriately scaling the background to the size of the source region (e.g., "EXPOSURE" and "BACKSCAL").



## You can also see what each individual component looks like in counts per keV, if you prefer:

```python
import glob
bgdfiles = glob.glob('bgd*pha')


plt.figure(figsize=(15, 10))
plt.loglog(data['CHANNEL']*0.04+1.6, data['COUNTS']/0.04, '+')
plt.ylim([1e-5, 1e5])
plt.ylabel("Counts per keV")
plt.xlabel("Energy (keV)")

for bgdfile in glob.glob('bgd*pha'):
#    print(bgdfile)
    
    bgd, bgd_hdr = getdata(bgdfile, 1, header=True)
    scale = (hdr['EXPOSURE'] * hdr['BACKSCAL']) / (bgd_hdr['EXPOSURE'] * bgd_hdr['BACKSCAL'])
 
    names = bgd.names
    if 'COUNTS' in names:
        scale = (hdr['EXPOSURE'] * hdr['BACKSCAL']) / (bgd_hdr['EXPOSURE'] * bgd_hdr['BACKSCAL']*0.04)
        plt.loglog(data['CHANNEL']*.04 +1.6, bgd['COUNTS']*scale)
    else:
        scale = (hdr['BACKSCAL']*hdr['EXPOSURE']) /  (0.04*bgd_hdr['BACKSCAL'])
        plt.loglog(bgd['CHANNEL']*0.04 + 1.6, bgd['RATE']*scale)

plt.show()
```

### Green in the Aperture CXB component, brown is the internal particle component, purple is the internal *activation* and fluorscence component, and pink is the combined model.

### One big caveat here is that all of this doesn't have all of the correction absorption parameters, so it will over-predict the background at low energies (below about 7 or 8 keV). This is something that we've been on the hook to fix in nuskybgd for a couple of years now, so maybe it'll happen.

# Below is how to deal with the RMF and ARF processing.

```python
def decompress_rmf(rmf):
    '''Code for decompressing the RMF that comes out of nuproducts 
    
    I'm sure there's a more pythonic way to do this, but...
    
    drmf[i][j] corresponds to the probability that a photons that arrives
    in the ith channel will be observed as a count in the jth channel of the
    spectrum.
    
    '''
    import numpy as np

    drmf = np.zeros([4096, 4096])
    for i in range(4095):
    
        compress_ind = 0
        start = rmf['F_CHAN'][i]
        width = rmf['N_CHAN'][i]
        for j in range(len(start)):
            drmf[i][start[j]:start[j]+width[j]] += rmf['MATRIX'][i][compress_ind:compress_ind+width[j]]
            compress_ind += width[j]
    
    # The transpose is necessary to make the RMF work with numpy's matmul method.
    return drmf.transpose()
            
```

```python
def fold_model(model, arf, rmf):
    '''Code for folding a model through the NuSTAR response functions.
    
    The "model" input should be a 4096 channel spectrum with the lowest edge starting at 1.6 keV and
    bin widths pf 0.04 keV. It's units should be ph / cm2 / sec (not **not** ph / cm2 / sec / keV)
    
    '''
    
    # Turns out this is a one-line module...
    foldrmf = np.matmul(rmf, arf*model)    
    
    return foldrmf

    
```

```python
# Extension 2 is the "MATRIX" extension that encodes the RMF
rmf = getdata('nu20211002001_sun_centerA_sr.rmf', 2)
rmf_header = getheader('nu20211002001_sun_centerA_sr.rmf', 2)
print(rmf.columns)

# We'll need to decompress it before we use it:
drmf = decompress_rmf(rmf)


# Extension 1 is the "ARF" extension. The "SPECRESP" field gives us the effective area as a function of energy.
arf = getdata('nu20211002001_sun_centerA_sr.arf', 1)
print(arf.columns)



    
    
```

```python
import numpy as np
# Let's make a toy model and see what happens
model = np.zeros(4096)
model[210] = 1.0 # 1 ph / cm2 at 10 keV

# Fold it through the response files.
model_cts = fold_model(model, arf['SPECRESP'], drmf)

f = plt.figure(figsize=(15, 10))
ax1 = plt.subplot(211)
#ax.set_xlim([0, 20])


ax1.plot(rmf['ENERG_LO'], model)
ax2 = plt.subplot(212, sharex=ax1)
ax2.plot(rmf['ENERG_LO'], model_cts, 'b')

ax1.set_xlim([0, 20])
ax1.set_xlabel("Energy (keV)")
ax1.set_ylabel("ph / cm2")
ax2.set_xlabel("Energy (keV)")
ax2.set_ylabel("Counts per channel")

plt.show()

```

---
## See how the blue (observed) curve is so much bigger than the cyan (incident) one?

### That's because of the effective area term in the ARF (which has units of cm2).

### This spectrum is now "expected counts per channel".

### **NOT** Counts per second per keV. To do that, simply divide by 40 eV.
